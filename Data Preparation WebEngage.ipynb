```python
import pandas as pd
```

```python
df = pd.read_csv('WebEngage_Bulk.csv')
df
```

```python
df.info()
```

```python
df['name'] = df['name'].apply(lambda x: x.title())
df['first_name'] = df['first_name'].apply(lambda x: str(x).title())
df['last_name'] = df['last_name'].apply(lambda x: str(x).title())
df.drop(['verified_email.1','customer_type','registration_status'],axis = 1, inplace = True)
df['phone'] = df.phone.apply(lambda x: '+971'+str(x))
df['phone.1'] = df['phone.1'].apply(lambda x: '+971'+str(x))
df['uuid'] = df.uuid.apply(lambda x: 'Customer_'+str(x))
df['email_subscription'] = df['email_subscription'].replace(True,'Subscribed')
df['email_subscription'] = df['email_subscription'].replace(False,'Unsubscribed')
df['promotional_sms_subscription'] = df['promotional_sms_subscription'].replace(True,'Subscribed')
df['promotional_sms_subscription'] = df['promotional_sms_subscription'].replace(False,'Unsubscribed')
df.geo_region_id = df.geo_region_id.astype('Int64')
df['created_at'] = pd.to_datetime(df['created_at'])
df['first_completed_order_date'] = pd.to_datetime(df['first_completed_order_date']).dt.strftime('%Y-%m-%d')
df['blacklisted_date'] = pd.to_datetime(df['blacklisted_date']).dt.strftime('%Y-%m-%d')
df['last_completed_order_date'] = pd.to_datetime(df['last_completed_order_date']).dt.strftime('%Y-%m-%d')
df['last_cancelled_order_time'] = pd.to_datetime(df['last_cancelled_order_time']).dt.strftime('%Y-%m-%d')
```

```python
df = df[['uuid', 'name', 'email', 'phone', 'email_subscription',
       'promotional_sms_subscription', 'created_at',
       'first_name', 'first_completed_order_date',
       'lto_count', 'promotional_sms_subscription.1', 'blacklisted', 'verified_email',
       'blacklisted_date', 'is_email_subscribed', 'last_name',
       'last_completed_order_date', 'reason_for_blacklisting', 'geo_region_id',
       'last_cancelled_order_time','email.1','phone.1']]
```

```python
df.columns = ['user_id', 'Name', 'Email', 'Phone', 'Has User Unsubscribed from Email?',
       'HasUserUnsubscribedfromSMS/IsUseronNDNCList?', 'Created On', 
       'first_name', 'first_completed_order_date', 
       'lto', 'is_promotional_sms_subscription', 'is_blacklisted', 'verified_email',
       'blacklisted_date', 'is_email_subscription', 'last_name',
       'last_completed_order_date', 'blacklisted_reason', 'geo_region_id',
       'last_cancelled_order_time','email','phone']
```
